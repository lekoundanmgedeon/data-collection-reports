### <font color = '#DEB887'> Pandas - Data Cleaning

```python
# import data.csv
df= pd.read_csv('data/data1.csv')
df.head()
```

```python
# Check the existence of null values 
df.isnull().sum()
```

```python
df.isnull().any().sum()
```

```python
df.isna().sum()
```

```python
df.isna().any().sum()
```

```python
# Delete null values using dropna 
df.dropna(inplace=True)
```

```python
df2 = pd.read_csv('data/data.csv')
df2
```

```python
for col in df2.columns[:2]: 
    df2[col].fillna(df2[col].median)
```

```python
df2.isnull().sum()
```

```python
# Replace null values by the mean
for col in df2.columns[:2]: 
    df2[col].fillna(df2[col].mean()0)
# import data.csv
```

```python
# Replace null values by the median 
for col in df2.columns[:2]: 
    df2[col].fillna(df2[col].median() )
# import data.csv

```

```python
# Replace null values by the mode 
# import data.csv
```

```python
# Check the types of the different variables 
```

```python
# Convert calories in integer using astype 
# import data1.csv
df.calories = df.calories.astype('int64')
df.dtypes
```

```python
# convert data in datetime64 using astype 
df.date = df.date.astype('datetime64[ns]')
df.dtypes
```

```python
# Delete rows (axis = 0) 

df.drop([0,1],axis=0)
```

```python
# Delete columns (axis = 1)
df.drop(['calories','duration'],axis=1)
```

```python
# Check the existence of duplicates 
# import data2.csv

df.duplicated()
```

```python
df.duplicated().sum()
```

```python
# Delete duplicates 
df.drop_duplicates()
```
