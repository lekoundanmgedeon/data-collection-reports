```python
import pandas as pd
from requests import get
from bs4 import BeautifulSoup as bs
```

```python
# url
url = 'https://dakar-auto.com/senegal/motos-and-scooters-3?&page=3'
#scrape the following informations:
#brand, model, year,reference, kms driven, adress, price, img_link
# from page 1 to 200
#clean the scraped data
```

```python
def scrape_page(page_number):
    url = f'https://dakar-auto.com/senegal/motos-and-scooters-3?page={page_number}'
    res = get(url)
    soup = bs(res.content, 'html.parser')
    # find all moto containers
    containers = soup.find_all('div', class_='listings-cards__list-item mb-md-3 mb-3')
    data = []
    for container in containers:
        try:
            # General info title
            title = container.find(
                'h2',
                'listing-card__header__title mb-md-2 mb-0'
            ).a.text.strip().split()
            brand = title[0]
            model = " ".join(title[1:-1])
            year = title[-1]

            # Details line info containers
            details = container.find_all(
                'li',
                'listing-card__attribute list-inline-item'
            )
            reference = details[0].text.split()[-1]
            kms_driven = details[1].text.replace(" km", "").strip()
            # Handle price
            price_raw = container.find(
                'h3',
                "listing-card__header__price font-weight-bold text-uppercase mb-0"
            ).text.strip()
            price = "".join(price_raw.split()).replace("FCFA", "")
            # Handle adresse and city
            addr_container = container.find("div", class_="entry-zone-address")
            if addr_container:
                addr_span = addr_container.find("span", class_="town-suburb")
                city_span = addr_container.find("span", class_="province")
                address = addr_span.text.replace(",", "").strip() if addr_span else None
                city = city_span.text.strip() if city_span else None
            else:
                address = None
                city = None
            # Image
            img_tag = container.find("img")
            img_link = 'https://dakar-auto.com/'+ img_tag["src"] if img_tag and img_tag.get("src") else None

            data.append({
                "brand": brand,
                "model": model,
                "year": year,
                "reference": reference,
                "kms_driven": kms_driven,
                "address": address,
                "city": city,
                "price": price,
                "img_link": img_link
            })
        except:
            pass

    return data
```

```python
df = pd.DataFrame()
for page in range(1, 201):
    print(f"Scraping page {page}...")
    page_data = scrape_page(page)
    #If page has no data, stop
    if len(page_data) == 0:
        break
    DF = pd.DataFrame(page_data)
    df = pd.concat([df, DF], axis=0).reset_index(drop=True)
print("Scraping completed!")
```

```python
# clean price
df["price"] = (
    df["price"]
    .astype(str)
    .str.replace(" ", "")
    .replace("NC", pd.NA)
)
df["price"] = pd.to_numeric(df["price"], errors="coerce").astype(float)
```

```python
# clean kms driven
df["kms_driven"] = (
    df["kms_driven"]
    .astype(str)
    .str.replace(" ", "")
)
df["kms_driven"] = pd.to_numeric(df["kms_driven"], errors="coerce")
```

```python
# clean year
df["year"] = pd.to_numeric(df["year"], errors="coerce")
```
