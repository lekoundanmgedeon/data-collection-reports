```python
url = 'https://www.basketball-reference.com/leagues/NBA_2010_per_game.html'
# Scrape the data tables from 2010 to 2024 using pandas.read_html and clean the data
```

```python
import pandas as pd
data = pd.read_html(url, header = 0)
len(data) # Let's check lenght of data of url
```

```python
#Use a for loop to  collect data from 2000 to 2024
df= pd.DataFrame()
for year in range(2000, 2025):
  url_page = f'https://www.basketball-reference.com/leagues/NBA_{year}_per_game.html'
  print(url_page) # Test si 2000-2024
  data = pd.read_html(url_page) # [df_page]
  df_page = data[0]
  df_page['year'] = year
  df = pd.concat([df, df_page], axis =0).reset_index(drop = True)
```

```python
# df dimension (just for checking, necessairly for  )
df.shape
```

```python
# Display of df
df.head()
```

```python
# 1.1 Check for null values
print(df.isnull().any().sum())
```

```python
# 1.2 Check for null values
print(df.isnull().sum())
```

```python
#2 Handle the non-values

# Replace by N/A for non numeric columns
df['Team'] = df['Team'].fillna("N/A")
df['Pos'] = df['Pos'].fillna("N/A")
df['Awards'] = df['Awards'].fillna("N/A")

# Replace all numeric (float and int) values by his mean
for col in df.columns:
    if df[col].dtype in ['float64', 'int64']:
        df[col].fillna(df[col].mean(), inplace=True)

# df = df.fillna(df.mean(numeric_only=True)) # or this panda method do also well
```

```python
#3. Check for duplicated value
print(df.duplicated().sum())
```

```python
# 4. Handle for duplicated values
df = df.drop_duplicates()
print(df.duplicated().sum())
# Nothing to do here, because sum of duplicated values is 0
```
